```python
#AI建立一個function
#資料來源:scores = np.random.randint(50,101,size=(50, 5))
#欄位:["國文", "英文", "數學", "自然", "社會", "總分","平均","排名"]
#索引:["學生" + str(i) for i in range(1, 51)]
#建立一個DataFrame
import numpy as np
import pandas as pd

def build_scores_dataframe(scores):
    df = pd.DataFrame(scores, columns=["國文", "英文", "數學", "自然", "社會"],
                      index=["學生" + str(i) for i in range(1, 51)])
    df.index.name = "學生"
    df.columns.name = "科目"
    sum_values = df.sum(axis=1)
    mean_values = df.mean(axis=1)
    df["總分"] = sum_values
    df["平均"] = mean_values
    rank_values = df["總分"].rank(method="min", ascending=False)
    df["排名"] = rank_values
    return df
scores = np.random.randint(50,101,size=(50, 5))
df = build_scores_dataframe(scores)
df
```

```python
#選擇欄位
df1 = df[["國文", "英文", "數學", "自然", "社會"]]
df1
```

```python
s = df1.loc["學生4"]
less_than_60_series = s[s < 60]
#取出不及格的科目
failed_subjects = less_than_60_series.index.tolist()
failed_subjects
#取出不及格的科目分數
failed_subject_scores = less_than_60_series.values.tolist()
failed_subject_scores
#建立一個Series,值要轉換為字串,若有多值則用逗號分隔
failed_subjects = ', '.join(failed_subjects)
failed_subject_scores = ', '.join(map(str, failed_subject_scores))
#建立一個Series,索引為["科目", "分數"]
result_series = pd.Series([failed_subjects, failed_subject_scores], index=["未及格科目", "分數"])
result_series
```

```python
def apply_statistics(s):
    less_than_60_series = s[s < 60]
    #不及格科目數量
    failed_subject_count = less_than_60_series.count()
    #取出不及格的科目
    failed_subjects = less_than_60_series.index.tolist()

    #取出不及格的科目分數
    failed_subject_scores = less_than_60_series.values.tolist()
    
    #建立一個Series,值要轉換為字串,若有多值則用逗號分隔
    failed_subjects = ', '.join(failed_subjects)
    failed_subject_scores = ', '.join(map(str, failed_subject_scores))
    return pd.Series([failed_subject_count, failed_subjects, failed_subject_scores], index=["不及格科目數量","未及格科目", "分數"])

df2 = df1.apply(apply_statistics,axis=1)
df2
```

```python
#df和df2合併
df_merged = pd.concat([df, df2], axis=1)
df_merged
```

```python
#df_merged儲存為excel檔
df_merged.to_excel("df_merged.xlsx")
```
