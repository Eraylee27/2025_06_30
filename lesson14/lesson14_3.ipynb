```python
#利用numpy和matplotlib產生正弦波和餘弦波
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)  # 產生從0到2π的100個點
y1 = np.sin(x)  # 計算正弦值
y2 = np.cos(x)  # 計算餘弦值

plt.plot(x, y2, label='Sine Wave(正弦波)')  # 繪製正弦波
plt.plot(x, y1, label='Cosine Wave(餘弦波)')  # 繪製餘弦波
plt.title('Sine Waves(正弦波) and Cosine Waves(餘弦波)')  # 圖表標題  
plt.xlabel('x (radians)')  # x軸標籤
plt.ylabel('y()')  # y軸標籤
plt.legend()  # 顯示圖例
plt.grid(True)  # 顯示網格
plt.show()  # 顯示圖表
```
